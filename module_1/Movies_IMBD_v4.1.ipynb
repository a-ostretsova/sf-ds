# 27. Какие актеры чаще всего снимаются в одном фильме вместе?


```python
answers['27'] = 'Daniel Radcliffe & Rupert Grint'
```

```python
def match_pairs(c):
    if ('Johnny Depp' in c) & ('Helena Bonham Carter' in c):
        return 'Johnny & Helena'
    elif ('Ben Stiller' in c) & ('Owen Wilson' in c):
        return 'Ben & Owen'
    elif ('Vin Diesel' in c) & ('Paul Walker' in c):
        return 'Vin & Paul'
    elif ('Adam Sandler' in c) & ('Kevin James' in c):
        return 'Adam & Kevin'
    elif ('Daniel Radcliffe' in c) & ('Rupert Grint' in c):
        return 'Daniel & Rupert'
    else:
        return 'None'

data['pairs'] = data['cast'].apply(match_pairs)
data['pairs'].value_counts()
```
